```python
import requests
from bs4 import BeautifulSoup
import sys
from collections import defaultdict, Counter
import pandas as pd
```

### Getting Batter Stats

```python
#batter url
batter_url = "http://www.rotowire.com/baseball/player_ex_stats.htm"
```

```python
def get_soup(url):
    r = requests.get(url)
    return BeautifulSoup(r.content, 'html.parser')
```

```python
batter_soup = get_soup(batter_url)
```

```python
#create dictionary of players and stats
def get_player_data(soup):
    tbl = soup.find('table')
    d = defaultdict(list)
    columns = []
    for idx, tr in enumerate(tbl.findAll('tr')):
        for td in tr.findAll('td'):
            d[idx].append(td.text)
    return d
```

```python
batter_dict= get_player_data(batter_soup)
```

```python
batter_data = pd.DataFrame.from_dict(batter_dict, 'index')
batter_data.head()
```

```python
#Manually retrieved columns from website
batter_data = batter_data.rename(columns={0: 'Player Name', 1:'Team', 2:'Position',
                                      3: 'Games', 4:'ABs',5: 'Pitches Per AB', 6:'Bunts',
                                     7: 'Hit into DP', 8:'Intentional Walks',9: 'Walks', 10:'SOs',
                                     11: 'Walk Rate', 12:'W2SO Ratio',13: 'Contact Rate', 14: 'Stolen Base Opportunity',
                                          15:'Batting Average',16: 'Batting Average on BIP'})
batter_data.head()
```

### Getting Pitcher Stats

```python
#pitcher url

pitcher_url = "http://www.rotowire.com/baseball/player_ex_stats.htm?pos=P"
pitcher_soup = get_soup(pitcher_url)
pitcher_dict = get_player_data(pitcher_soup)
pitcher_data = pd.DataFrame.from_dict(pitcher_dict, 'index')
pitcher_data.head()
```

```python
pitcher_data = pitcher_data.rename(columns={0: 'Player Name', 1:'Team',2: 'Games', 3:'Innings Pitched',
                                      4: 'Starts', 5:'Quality Starts',6: 'Quality Start %', 7:'K/BB',
                                     8: 'K/9IP', 9:'BB/9IP',10: 'HR/9IP', 11:'Batting Avg on Balls IP',
                                     12: 'Strand Rate', 13:'FB VLO',14: 'ERA', 15:'Fielding Ind Pitching',
                                     16: 'Batting Average on BIP'})
pitcher_data.head()
```
